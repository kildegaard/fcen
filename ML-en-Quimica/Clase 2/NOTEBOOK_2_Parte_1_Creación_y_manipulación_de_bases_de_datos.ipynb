## Principios FAIR

Algunas de las primeras preguntas que hay que hacerse al crear una base de datos son:
* Si alguien externo a mi grupo de trabajo (o usted mismo en un futuro) quiere usar mi base de datos, ¿va a entenderla?
* ¿Se entiende que significa cada variable?
* ¿Se conocen las unidades de los datos? 
* ¿Se explican las abreviaciones y los términos técnicos?

Una manera de asegurarse de que mi base de datos sea reutilizable es seguir los lineamientos internacionales llamados *principios FAIR* como guía. 

Los principios FAIR para la gestión y administración de datos científicos ([artículo](#https://doi.org/10.1038/sdata.2016.18)) se desarrollaron como directrices para mejorar la capacidad de búsqueda, la accesibilidad, la interoperabilidad y la reutilización de los activos digitales, todo lo cual apoya la reproducibilidad de la investigación.

La idea general es que los datos FAIR deben ser:

* Localizables (*Findable*): Es importante poder encontrar los datos para utilizarlos.

* Accesibles (*Accessible*): Luego de encontrarlos, se debe poder acceder a ellos (podrían ser o no de libre acceso).

* Interoperables (*Interoperable*): Los datos deben integrarse con otros datos e interoperar con aplicaciones o flujos de trabajo. Por ejemplo, quisiera poder abrirlos en Python.

* Reutilizables (*Reusable*): Los datos deben estar bien descritos para que puedan ser utilizados, combinados y ampliados en diferentes entornos.

Para más información, recomendamos leer https://the-turing-way.netlify.app/reproducible-research/rdm/rdm-fair.html 


## Documentación y metadata
Los datos que recolectemos deben poder ser entendidos por otra persona o grupo de trabajo. Por lo tanto, hay que asignarles un contexto, una descripción.

Para lograr este objetivo es importante tener **documentación** y **metadata**


*   La **documentación** brinda el contexto: para qué se buscaron los datos, cuáles son las fuentes, qué limitaciones tienen, quien creó la base de datos, cuándo, cuál fue la última modificación. También saber si es open access o si se necesitan permisos para modificarla. 

*   La **metadata** es la información sobre los datos de tal manera que éstos queden totalmente descriptos: explicar qué es cada variable, si son numéricas o categóricas, qué unidades tienen, etc. 

Por ejemplo, cuando en un archivo hacemos click derecho --> propiedades, estamos viendo la metadata. Lo mismo con una foto del celular.

## Organización de datos en spreadsheets
Microsoft Excel y Google Sheets son las herramientas más utilizadas para el análisis de datos. Permiten organizar la información en forma de tabla donde:


*   Las columnas son las variables

*   Las filas son las observaciones 

*   Las celdas son los valores

![picture](https://drive.google.com/uc?export=view&id=1uQl3Y7s9ojqJpvs1VNoEUVUErBc4e6wx)

Existe **una lista de cosas que NO se deben hacer** al trabajar con spreadsheets pues pueden ocasionar errores al exportar los datos:

*   Cambiar fuente y/o color de la letra

*   Usar funciones

*   Combinar celdas

*   Usar formato fecha en las celdas (es mejor armar 3 columnas; año, mes y día)

*   Manipular y analizar los datos (incluido el copy-paste): lleva a workflows no reproducibles

Por otro lado, **las cosas que SÍ se deben hacer** (o son muy recomendables) son:


*   Ser consistentes a la hora de ingresar los datos. Por ejemplo, en un ensayo interlaboratorio, la fecha 02-03 será 3 de febrero en USA, pero 2 de marzo en Latinoamérica.

*   Ser consistentes con los missing values. Las celdas pueden quedar vacías, NULL o NA y cada una puede significar algo distinto. Por ejemplo, vacía es un dato que se perdió, NULL un dato que todavía no se midió y NA un dato imposible de medir.

*   Guardar los datos en formato texto, ya sea .csv o .tsv

*   Si trabajamos con varias sheets en el mismo archivo (formato xlsx), dado que el formato csv permite guardar una sheet a la vez resulta conveniente guardar cada sheet en csv por separado y unirlas luego en un archivo .zip

*   Dado que los datos se recogen de forma manual, es conveniente escribirlos en el mismo formato en el que van a ser ingresados a la sheet en una versión impresa de la tabla. Esto permite automatizar el ingreso de la información y evita errores humanos (https://support.microsoft.com/en-us/office/insert-data-from-picture-3c1bb58d-2c59-4bc0-b04a-a671a6868fd7#ID0EBBD=iOS/Android)


*   Al trabajar en equipo, es recomendable que todos tengan la misma versión del software a utilizar y que estén claros los modos de ingresar los datos.

La información descripta en bases de datos puede ser abierta. Es decir, puede encontrarse disponible en Internet para que cualquier usuario pueda descargarla, copiarla, analizarla, reprocesarla y reutilizarla para cualquier otro fin, con un mínimo de barreras financieras, legales y técnicas.

### Privacidad y protección de datos
Que los datos se encuentren abiertos no significa que se pierda su propiedad. Es importante siempre aplicar una **licencia** a los datos. Se trata de un documento legal que establece los permisos para el trabajo creativo y académico. Explica los derechos de autor, garantiza la atribución adecuada y establece cómo otros pueden copiar, distribuir y hacer uso de las obras.

Muchos campos de las disciplinas científicas implican trabajar con datos personales sensibles, por ejemplo, medicina. Su gestión está bien regulada en la legislación sobre protección de datos y en los procedimientos éticos establecidos en la mayoría de las instituciones de investigación.

## Publicar bases en artículos
Un artículo de datos (*Data Article*) es una publicación centrada en la descripción de un conjunto de datos. Utiliza la estructura tradicional de un artículo de revista, pero se centra en la recogida de datos y en los aspectos metodológicos y, por lo general, no en la interpretación o discusión de los resultados.

El gran beneficio de un artículo es que su producción será revisada por pares, algo que generalmente no ocurre con los datos que se archivan en los repositorios. La publicación de un artículo de estas características aumentará, por tanto, la visibilidad, la credibilidad y la utilidad de los datos, además de darle crédito al productor de tales datos.

Existen muchas revistas especializadas en la publicación de artículos de datos, como [Scientific Data](#https://www.nature.com/sdata/), [Data in Brief](#https://www.sciencedirect.com/journal/data-in-brief), entre otras.

## Concatenar dataframes




La función para concatenar dataframes. Teniendo en cuenta que tenemos una base de datos de vinos tintos y blancos, resulta de interés fusionarlos y agregar una variable categórica *hue*

## Tablas pivot

Resulta de interés teniendo 2 variables categóricas, *hue* y *quality*, hacer una pivot table al igual que en Excel. 

¿Qué es una pivot table?

Es una herramienta que permite analizar y resumir grandes cantidades de datos de manera rápida y sencilla. Se utiliza para transformar datos de una tabla o de una consulta de base de datos de forma que se puedan ver de diferentes maneras, por ejemplo, para ver los datos agrupados por categorías.

Para ello, necesitamos definir:

*   Las filas o el índice
*   Las columnas
*   Los valores que pueden ser promedios, valores máximos, mínimos, sumas, etc.




```python
pivot_table = df_wine.pivot_table(index='quality', columns='hue', values='alcohol')
pivot_table
```

Si bien la tendencia no es del todo uniforme, pareciera que a mayor contenido de alcohol la calidad es mayor. 

En el caso de que no quisiéramos ver el promedio, se puede modificar el parámetro *aggfunc*

```python
pivot_table = df_wine.pivot_table(index='quality', columns='hue', values='alcohol', aggfunc = "count")
pivot_table
```

Se observa que la mayoría de vinos, tanto para lo tintos como para los rojos, tienen calidades entre 5 y 6. Podemos relativizar estos valores en función de la cantidad de vinos de cada tipo:


```python
pivot_table["red%"] = (pivot_table["red"]/1599)*100
pivot_table["white%"] = (pivot_table["white"]/4898)*100
pivot_table
```

Haciendo esto se observa una tendencia y es que los vinos blancos parecieran tender a tener una mejor calidad que los vinos tintos.

## Estadística con groupby





Algo parecido a las pivot tables, pero no exactamente igual, consiste en agrupar valores en base a una categoría y mirar parámetros estadísticos como promedios, máximos, mínimos, etc. Para ello utilizamos la función *group_by*:

```python
df_wine.groupby(["hue"]).mean()
```

Rápidamente, vemos que no es tan claro que los vinos blancos tengan más calidad que los tintos como habíamos supuesto antes. Sin embargo, para estar completamente seguros habría que:

## Personalización de tablas

Una posibilidad es que se resalten valores con un color en específico. Por ejemplo:

```python
df_manual = pd.DataFrame({"Columna1": np.array([1,2,3, np.nan, 5, 1]), "Columna2": np.array([6,np.nan,8,9,10, 6]), "Columna3": np.array([11, 12, 13, 14, np.nan, 11]), 
                          "Columna4": np.array([7, 50, 9, 5, 1, 7])})

df_manual.style.highlight_min(color = "violet").highlight_max(color = "green").highlight_null()
```

```python
df_manual.style.background_gradient()
```

Por último, cambiar fondo, bordes y letra:

```python
df_manual.style.set_properties(**{'background-color': 'green', 'border': 'solid blue'}, color="white")
```
