## Breve introducción al algoritmo (opcional)

Un árbol de decisión es una técnica no lineal que divide el conjunto de datos en subconjuntos según el valor de una característica. Este paso se repite de forma recursiva en cada nuevo subconjunto hasta que se alcanzan algunos criterios. Por ejemplo, cada punto del subconjunto tiene el mismo valor de la variable objetivo o se alcanza una profundidad máxima. La elección de la característica y el valor a dividir se basa en reducir la entropía objetivo dividiéndola en subconjuntos puros. Para obtener más información, visite la [documentación](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html 'sklearn.tree.DecisionTreeClassifier').

La nomenclatura utilizada para referirse a cada parte del árbol se puede ver [acá](https://www.ibm.com/topics/decision-trees#:~:text=A%20decision%20tree%20is%20a,internal%20nodes%20and%20leaf%20nodes.) (el ejemplo de la derecha es de un dataset de flores):

image.png

¿A qué nos referimos cuando hablamos de entropía? Como químicos lo primero que se nos puede venir a la mente es el segundo principio de la Termodinámica. Y esto no está del todo mal si lo miramos desde el punto de vista de la termodinámica estadística.

El principio de Boltzmann establece que la entropía, S, es una constante k multiplicado por el logaritmo natural del número de microestados, Ω.

image.png

¿Qué son los microestados? Dar una definición formal carece de utilidad en este contexto pero podemos decir que se relacionan con el número de configuraciones que un sistema puede optar. 

Podemos ver Ω como una medida de nuestra falta de conocimiento sobre un sistema. Como ilustración de esta idea, considere un conjunto de 100 monedas, cada una de las cuales es cara o cruz. Los macroestados se especifican por el número total de caras y cruces, mientras que los microestados se especifican por la configuración de cada moneda individual. Para los macroestados de 100 caras o 100 cruces, existe exactamente una configuración posible, por lo que nuestro conocimiento del sistema está completo. En el extremo opuesto, el macroestado que nos proporciona el menor conocimiento sobre el sistema consta de 50 caras y 50 cruces en cualquier orden, para las cuales hay 100.891.344.545.564.193.334.812.497.256 (número combinatorio (100,50)) ≈ 10^29 posibles microestados.



Entonces, si queremos tener la máxima información posible del sistema, debemos asegurarnos de que Ω sea chico. Esto equivale a pedir que la entropía sea chica, S. No es de extrañar entonces que en un árbol de decisión, el split o división del dataset busque MINIMIZAR la entropía, o lo que es lo mismo, minimizar la pérdida de información.
