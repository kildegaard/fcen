<u>Importancia de la toma de decisiones:</u>
<br>
- Materia prima con la cual construimos nuestras vidas
- Estar donde uno está es a causa de la toma de deciciones realizada a lo largo del tiempo

<u>Cambios en la forma de tomar decisiones:</u>
- Baja el costo de la adquisición y procesamiento de los datos
- Paso a tener muchos datos -> escasez de tiempo para analizar esa info

Con más información para tomar decisiones:
- Decisiones intuitivas --> **decisiones cuantitativas!**

Decisiones:
- Incrementar las operaciones (vale la pena operar a un mayor nivel o aumentar la inversión)
- Optimizar operaciones para un dado proceso (tratar de operarlo de la mejor manera posible)

Cómo se suele manejar esto?
- Incrementar operaciones (cuando te va bárbaro)
- Optimizar operaciones (cuando te está yendo más o menos)

### Tipo de decisiones

### Niveles de sofistificación para la toma de decisiones

- Programación lineal
  - Simplificaciones en la toma, tipo ecuaciones bobas, "tipo excel"
  - Útil cuando vos podés controlar todas las variables (si vos generás todo)
- Modelos rigurosos
  - Cosas más complejas que tienen en cuenta muchas variables
  - Útil cuando no tenés ese control, por ejemplo cuando comprás cosas afuera y tenés una alta variabilidad en tus condiciones

### Blending automático

Quiero una nafta de una condición en particular, por ejemplo el mejor octanaje, mezclando distintas fuentes o tanques

- Puedo mezclar todo y medir luego "qué octanaje obtuve"
- O puedo medir online e ir corrigiendo minuto a minuto

Midiendo el NIR y aplicando DS uno **infiere** el valor del octanaje puesto que no puedo medir "rápido y online" el mismo

Ahora, cómo lo vendés? Por qué es mejor?

Se ve que con DS la toma de datos nos ahorra mucha plata. El octanaje obtenido es mucho más cercano al target que *hacerlo a ojo*. Ahí le estás regalando plata pq **vos se lo vendés como 97 octanos** cuando en realidad es 99 en promedio!!
