## Clustering

<u>Agrupar datos por similitud</u>

- Quiero predecir una propiedad. Tomo todos los datos para tratar de predecir con un modelo complejo... O puedo subdividir en conjuntos más pequeños y hacer modelitos más simples para cada uno de esos grupos.
- Permite encontrar relaciones locas
- No hace falta saber "qué estoy buscando", osea, el ser **no supervisado** hace que los clusters se generen solos

- Necesito menos conocimiento del sistema, es díficil encontrar relaciones a ojos.
- Peligroso: cuando tengo pocos datos! Corro el peligro de sobreajustar

Puedo usar muchos tipos de **distancia**.

Agrupo los datos por cercanía:
- Distancia euclídea
- Mahalanobis
- Euclídea cuadrado...

### 1er Método de clusterización - K Means

El algoritmo empieza poniendo en el espacio la cantidad de centroides que pensamos que va a haber

1. Elijo k puntos al azar
2. Asigno cada individuo al centro más cercano con un color
3. Calculo el centro de masa de cada grupo y muevo allí el centro
4. Repito desde 2 hasta cambio mínimo!

### 2do Método de clusterización - Clustering Jereárquico

- Comenzamos con tantos clústers como puntos
- Luego veo los 2 puntos que más cerca están entre sí: pasan a formar un grupo y los reemplazo por un centroide entre ellos
- Repito. Voy formando clusters

Armo **dendogramas** que me permiten ver la info de una manera copada

![Dendogramas](img03/img2.png)

Ventajas y desventajas:
- Siempre se obtiene la misma solución
- Nos da automáticamente todas las combinaciones posibles entre clústers
- Es más sensible a outliers (leve desventaja)
- No funciona bien con muchos datos

## Principal Component Analysis

Usos:
- Reducción de dimensionalidad
- Diagnóstico de fallas

Agarra los ejes coordenados en $R^{nxm}$ canónicos y los rota: genera nuevos ejes.

Ventajas:
- Coinciden con las principales direcciones de varianza del sistema (sus autovectores)
- Lo asocio a una reducción de la dimensionalidad eligiendo los autovectores con autovalores más grandes

Reduzco dimensión pues, por ejemplo, si tengo puntos que se agrupan en una recta y=x y roto los ejes X e Y para coincidir con esto, todos los datos están "en X" y poquito en Y, con lo cual puedo quedarme con X, descartar Y **sin perder demasiada información**.

Los autovalores me van a dar los pesos de los autovectores!

Cómo selecciono la cantidad de Componentes Principales? (CP)

Tengo que ver de quedarme con 2 o 3 componentes (para poder graficar) tratando de tener un buen % explicado (sabiendo que será menor al 100%)

[Hay muchos ejemplos en las diapos]

PCA se parece bastante a la regresión...

Diferencias:
- Regresion: busca el hiperplano que minimiza la distancia, en la dirección de la Y canónica, de cada punto a dicho hiperplano
- PSA: busca minimizar la distancia **perpendicular** al hiperplano

- Si tenés muchos puntos y están bien correlacionados, el plano que uno va a obtener se ve claramente y no hay mucha diferencia entre minimizar ambas distancias (regresión o psa).
- Desde el punto de vista numérico sí hay un tema!
- Cuando se hace regresión, tengo que invertir la matriz de covarianza.
  - Si hay variables en el dominio que correlacionan mucho entre sí, eso hace que la matriz de covarianza sea singular, que no sea invertible y trae muchos problemas de estabilidad numérica en la regresión.

**NOTA:** Si no podés garantizar la singularidad de la matriz, usá PSA. Sino andá por regresión que es de los mejorcitos.
